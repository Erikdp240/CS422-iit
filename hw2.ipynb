### **2.1 Problem 1**

```python
# importing stuff
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as sk
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules

path = r"https://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx"
df = pd.read_excel(path)

france = df["Country"] == "France"
france_df = df[france][["InvoiceNo", "Description"]]
france_df[["InvoiceNo", "Description"]]

col = france_df["Description"].unique()
row = france_df["InvoiceNo"].unique()

txn_df = pd.DataFrame(columns=col)
txn_df.insert(column="InvoiceNo", loc=0, value=row)
txn_df = txn_df.replace(np.nan, False)

for idx, row in france_df.iterrows():
    invoice, item=row["InvoiceNo"], row["Description"]
    txn_df.loc[txn_df["InvoiceNo"] == invoice, item] = True

txn_df = txn_df.drop("InvoiceNo", axis=1)

frequency_sets = apriori(txn_df, min_support=0.05, use_colnames=True)

rules_conf = association_rules(frequency_sets, metric="confidence", min_threshold=1)

rules_lift = association_rules(frequency_sets, metric="lift", min_threshold=11)
rules_lift
```

* the rule with the highest confidence is not the same as the highest lift.
* given the antecedent, confidence measures how often the rule is correct.
* lift measures the strength of the association between the antecedent and consequent.

### **2.2 Problem 2**

```python
path = r"75000-out2-binary.csv"
bakery_df = pd.read_csv(path)

name_1 = "Chocolate Coffee"
name_2 = "Chocolate Cake"

cake_coffee = bakery_df[[name_1, name_2]]

cond_1 = cake_coffee[name_1] == 1
cond_2 = cake_coffee[name_2] == 1

stats = cake_coffee.groupby([cond_1, cond_2]).count()
stats
```

```python
print(cake_coffee["Chocolate Coffee"].corr(cake_coffee["Chocolate Cake"]))
print(cake_coffee["Chocolate Cake"].corr(cake_coffee["Chocolate Coffee"]))
```

* the two items are symmetric binary variables.
* the association rules would have the same correlation values.
