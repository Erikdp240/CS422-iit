### **2.1 Problem 1**

```python
import seaborn as sn
```

```python
# Import data 'titanic'
data = sn.load_dataset('titanic')
data
```

```python
# Only those who survived
survived = data[data.survived==1]
```

```python
# Histogram of survival and gender
survived['sex'].hist(bins=3)
```

**What is the basic relationship between these variables using visual inspection? Do the results make sense? Why?**

The basic relationship between these variables is that females had survived more than males did. These stats make sense because in the Titanic, females and babies were put in lifeboats first then males.

```python
# Histogram of survival and age
survived['age'].hist(bins=60)
```

**What is the basic relationship between these variables using visual inspection? Do the results make sense? Why?**

The basic relationship between these variables is that people in their 20s or 30s survived more than other ages. These stats makes sense because in the Titanic, people who were on board they were around those ages.
