$$

\begin{pmatrix}
x_1^T \\
x_2^T \\
... \\
x_n^T
\end{pmatrix}
$$

$$
mean(D) = µ = \frac{(x_1^T + x_2^T + ... + x_n^T)}{n}
$$

$$
(1) \\
n * µ = (x_1^T + x_2^T + ... + x_n^T)
$$

$$
mean(\bar D) = \frac{(x_1^T-µ^T) + (x_2^T-µ^T) + ... + (x_n^T-µ^T)}{n}
$$

$$
mean(\bar D) = \frac{(x_1^T + x_2^T + ... + x_n^T) - (n*µ^T)}{n}
$$

$$
(1) \\ 
(x_1^T + x_2^T + ... + x_n^T) = n*µ
$$

$$
mean(\bar D) = \frac{n*µ - n*µ}{n}
$$

$$
mean(\bar D) = \frac{0}{n} = 0
$$

$$
\Gamma(\alpha) = \int_{0}^\infty x^{\alpha - 1}e^{-x} dx
$$

### **(a)**
$$
\Gamma(1) = \int_{0}^\infty x^{1- 1}e^{-x} dx = \int_{0}^\infty x^{0}e^{-x} dx = \int_{0}^\infty e^{-x} dx = [-e^{-x}]_{0}^\infty
$$

$$
\lim_{x \to \infty} -e^{-x} = -\frac{1}{e^x} = -\frac{1}{e^\infty} = -\frac{1}{\infty} = 0
$$

$$
\Gamma(1) = 0 - (-e^0) = 0 - (-1) = 1
$$

$$
\Gamma(1) = 1
$$

### **(b)**
$$
\Gamma(\frac{1}{2}) = \int_{0}^\infty x^{\frac{1}{2}- 1}e^{-x} dx = \int_{0}^\infty x^{-\frac{1}{2}}e^{-x} dx = \int_{0}^\infty \frac{e^{-x}}{\sqrt{x}} dx
$$

$$
u = \sqrt{x}
$$

$$
du = \frac{dx}{2\sqrt{x}}
$$

$$
\Gamma(\frac{1}{2}) = 2*\int_{0}^\infty e^{-u^2} du = \int_{-\infty}^\infty e^{-u^2} du
$$

$$
\Gamma(\frac{1}{2}) = \sqrt{\pi}
$$

$$

```python
import numpy as np
import matplotlib.pyplot as plt

mean = np.array([6, -12])
covariance = np.array([[22, -47.5], [-47.5, 110.5]])

# Generate points on a grid
x, y = np.meshgrid(np.linspace(-20, 30, 100), np.linspace(-40, 16, 100))
pos = np.dstack((x, y))

# Calculate the PDF using multivariate normal distribution
delta = pos - mean
inv_cov = np.linalg.inv(covariance)
pdf = (1 / (2 * np.pi * np.sqrt(np.linalg.det(covariance)))) * np.exp(-0.5 * np.einsum('...k,kl,...l->...', delta, inv_cov, delta))

# Plot the contour of the PDF
plt.contourf(x, y, pdf, cmap='viridis')


plt.xlabel('X1')
plt.ylabel('X2')
plt.title('2D Normal Density Function')

plt.show()
```
