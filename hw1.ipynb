$$

$$

$$

$$

### **(a)**
$$

$$

$$

$$
\Gamma(1) = 0 - (-e^0) = 0 - (-1) = 1
$$

$$
\Gamma(1) = 1
$$

$$

$$
u = \sqrt{x}
$$

$$
du = \frac{dx}{2\sqrt{x}}
$$

$$
\Gamma(\frac{1}{2}) = 2*\int_{0}^\infty e^{-u^2} du = \int_{-\infty}^\infty e^{-u^2} du

### **2.2 Problem 2**

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
# Load dataset into Pandas dataframe
path = 'https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data'
names = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'car name']
df = pd.read_csv(path, sep='\s+', names=names, na_values='?')
summary_stats = df.describe()
summary_stats
```

```python
# Replace missing values in 'horsepower' with mean
im_mean = SimpleImputer(strategy='mean')
df_mean = df.copy()
df_mean['horsepower'] = im_mean.fit_transform(df[['horsepower']])
```

```python
# Replace missing values in 'horsepower' with median
im_median = SimpleImputer(strategy='median')
df_median = df.copy()
df_median['horsepower'] = im_median.fit_transform(df[['horsepower']])
```

```python
# Replace missing values in 'horsepower' with mode
im_mode = SimpleImputer(strategy='most_frequent')
df_mode = df.copy()
df_mode['horsepower'] = im_mode.fit_transform(df[['horsepower']])
```

```python
# Calculate the variance of 'horsepower'
var_mean = df_mean['horsepower'].var()
var_median = df_median['horsepower'].var()
var_mode = df_mode['horsepower'].var()
```

```python
# Find the lowest varaince method and value
imp = ['mean', 'median' 'mode']
var = [var_mean, var_median, var_mode]

min_var_value = min(var)
min_var_method = imp[var.index(min_var_value)]

print(f"Imputation method with the lowest variance: {min_var_method}")
print(f"Variance value: {min_var_value}")
```

##### The imputation that results in the lowest variance is **mean**. This is because imputing the mode on the missing values of 'horsepower' creates less amount of variability in the data.

### **2.3 Problem 3**

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

# Load iris into Pandas dataframe
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# PCA on dataset
pca = PCA()
pca.fit(df)

# Variance explained
var_ex = pca.explained_variance_ratio_
var_feat = df.var() / df.var().sum()

# Print results
print("Percentage of Variance Explained by Principal Components:")
for i, var in enumerate(var_ex):
    print(f"Principal Component {i+1}: {var*100:.2f}%")

print("\nPercentage of Variance Explained by Original Features:")
for i, var in enumerate(var_feat):
    print(f"Feature {i+1}: {var*100:.2f}%")
```

##### The original features had **low** varainces

### **2.4 Problem 4**

```python
import matplotlib.pyplot as plt

# Get projections
pc1 = pca.components_[0]
proj = np.dot(df.values, pc1)

# Plot projections
fig, axs = plt.subplots(2, 2, figsize=(12,8))
axs = axs.flatten()

for i, feature in enumerate(df.columns):
    axs[i].scatter(df[feature], proj)
    axs[i].set_xlabel(feature)
    axs[i].set_ylabel("Projection onto PC1")

plt.tight_layout()
plt.show()

# Calculate the correlation coefficient between the selected pair
selected_features = ['sepal length (cm)']
corr_coef = np.corrcoef(df[selected_features].values.T, proj)[0, 1]
print(f"Correlation coefficient (sepal length v. PC1): {corr_coef:.4f}")

selected_features = ['sepal width (cm)']
corr_coef = np.corrcoef(df[selected_features].values.T, proj)[0, 1]
print(f"Correlation coefficient (sepal width v. PC1): {corr_coef:.4f}")

selected_features = ['petal length (cm)']
corr_coef = np.corrcoef(df[selected_features].values.T, proj)[0, 1]
print(f"Correlation coefficient (petal length v. PC1): {corr_coef:.4f}")

selected_features = ['petal width (cm)']
corr_coef = np.corrcoef(df[selected_features].values.T, proj)[0, 1]
print(f"Correlation coefficient (petal width v. PC1): {corr_coef:.4f}")
```

##### The numerical results match the visual inspection because the closer the coefficient is to 1 then the stronger visual correlation there is.

### **2.5 Problem 5**

```python
# Calculate total variance of original features
total_variance_og = df.var().sum()

# Calculate total variance explained by four eigenvectors
total_variance_pca = sum(pca.explained_variance_)

# Determine the number of Principal Components required to capture >95%
cumulative_variance_ratio = np.cumsum(pca.explained_variance_ratio_)
n_components = np.argmax(cumulative_variance_ratio > 0.95) + 1

# Determine number of dimensions
n_dimensions = len(pca.components_)

# Print
print(f"Total variance of original features: {total_variance_og:.4f}")
print(f"Total variance explained by Principal Components: {total_variance_pca:.4f}")
print(f"Number of Principal Components required to capture > 95% of the variance: {n_components}")
print(f"Number of dimensions reduced to: {n_dimensions}")
```
